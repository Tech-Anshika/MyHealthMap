```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd

# Load the dataset
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')

# Selecting features and target
X = df[['Sleep Duration', 'Physical Activity Level']]
y = df['Heart Rate']

# Train-test split (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Train Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mean_sq_er = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('Mean Squared Error:', mean_sq_er)
print('R2 Score:', r2)

# Predict for a new input: Sleep Duration=6.5 hours, Physical Activity=45 minutes
print('Predicted Heart Rate:', model.predict([[6.5, 45]]))
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd

# Load the dataset
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')

# Select features and target
X = df[['Sleep Duration', 'Physical Activity Level']]
y = df['Heart Rate']

# Train-test split (90% training, 10% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Initialize and train SVR model
svm_model = SVR(kernel='rbf')  # Options: 'linear', 'poly', 'rbf'
svm_model.fit(X_train, y_train)

# Make predictions
y_pred = svm_model.predict(X_test)

# Evaluation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R2 Score:", r2)

# Predict Heart Rate for a new person (Sleep Duration = 6.5 hrs, Physical Activity = 45 mins)
print("Predicted Heart Rate:", svm_model.predict([[6.5, 45]]))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
import pandas as pd

# Load dataset
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')

# Create heart rate category labels
def heart_rate_category(hr):
    if hr < 60:
        return "Low"
    elif hr <= 100:
        return "Normal"
    else:
        return "High"

df['HeartRateCategory'] = df['Heart Rate'].apply(heart_rate_category)

# Features and target
X = df[['Sleep Duration', 'Physical Activity Level']]
y = df['HeartRateCategory']

# Encode target labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.1, random_state=42)

# --- Random Forest Classifier ---
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
rf_acc = accuracy_score(y_test, rf_model.predict(X_test))
print("Random Forest Accuracy:", rf_acc)

# --- K-Nearest Neighbors ---
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)
knn_acc = accuracy_score(y_test, knn_model.predict(X_test))
print("KNN Accuracy:", knn_acc)

# --- Neural Network (MLPClassifier) ---
nn_model = MLPClassifier(hidden_layer_sizes=(50,), max_iter=1000, random_state=42)
nn_model.fit(X_train, y_train)
nn_acc = accuracy_score(y_test, nn_model.predict(X_test))
print("Neural Network Accuracy:", nn_acc)
```

```python
from flask import Flask, request, jsonify
import joblib
import numpy as np
import pandas as pd
import threading

# Create Flask app
app = Flask(__name__)

# Load your model (ensure this file exists in the same directory)
model = joblib.load('sleep_disorder_model.pkl')  # Your trained model

# Prediction route
@app.route('/predict', methods=['POST'])
def predict():
    data = request.get_json()

    try:
        features = [
            data['Age'],
            1 if data['Gender'] == 'Male' else 0,
            float(data['Sleep Duration']),
            float(data['Quality of Sleep']),
            float(data['Physical Activity Level']),
            float(data['Stress Level']),
            float(data['Heart Rate']),
            float(data['Daily Steps']),
            float(data['Caffeine Consumption']),
            float(data['Alcohol Consumption']),
            float(data['Smoking Frequency'])
        ]

        prediction = model.predict([features])[0]
        disorder_map = {0: "Insomnia", 1: "None", 2: "Sleep Apnea"}
        return jsonify({'prediction': disorder_map.get(prediction, 'Unknown')})

    except Exception as e:
        return jsonify({'error': str(e)})

# Run Flask in a background thread (Jupyter Notebook safe)
def run_flask():
    app.run(debug=True, use_reloader=False)

thread = threading.Thread(target=run_flask)
thread.start()
```
